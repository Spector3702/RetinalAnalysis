```python
import cv2
import os
import random
import shutil
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/Training_Set/RFMiD_Training_Labels.csv')
df.columns
```

```python
def plotImages(images):
    sampled_ids = random.sample(list(images), min(15, len(images)))

    plt.figure(figsize=(15, 10))
    for i, img_id in enumerate(sampled_ids, 1):
        img_path = os.path.join('data/Training_Set/Training', str(img_id) + '.png')
        img = cv2.imread(img_path)
        if img is not None:
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
            plt.subplot(3, 5, i)
            plt.imshow(img_rgb)
            plt.title(img_id) 
            plt.axis('off')
```

```python
def copy_files(filenames, destination_folder):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for filename in filenames:
        if os.path.exists(filename):
            shutil.copy(filename, destination_folder)
        else:
            print(f"File {filename} not found.")
```

## List TV Sample Images

```python
tv_images = df[df['TV'] == 1]['ID']
tv_images
```

```python
plotImages(tv_images)
```

```python
filnames = [
    'data/Training_Set/Training/18.png',
    'data/Training_Set/Training/357.png',
    'data/Training_Set/Training/907.png',
    'data/Training_Set/Training/915.png',
    'data/Training_Set/Training/1329.png',
    'data/Training_Set/Training/1530.png'
]
destinationFolder = 'data/Sample_imgs/TV'
copy_files(filnames, destinationFolder)
```
